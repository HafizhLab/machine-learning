# Load Corpus

```python
with open('alquran_korpus.txt') as f:
    ayahs = f.readlines()
```

# Sambung Kata

## Train Model

```python
from lm.markovchain import MarkovChain

# load model
model = MarkovChain()

# train
for ayah in ayahs:
    model.add_document(ayah)
```

```python
# test model وَاللَّهُ
model.next_word(input())
```

```python
# test model وَاللَّهُ بِمَا تَعْمَلُونَ
model.next_word(input())
```
